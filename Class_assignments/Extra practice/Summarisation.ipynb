# **Visualization of Data Distribution**

```python
df= pd.read_csv('https://raw.githubusercontent.com/jbryer/CompStats/master/Data/titanic3.csv')
df.head(3)
```

```python
df.shape
```

**Histogram**

```python
import matplotlib.pyplot as plt
ages=df['age'].dropna()
plt.hist(ages, bins=20) # We can change the bins size by increasing or reducing the value
plt.xlabel('Ages (years)');
plt.ylabel('Number of Passengers');
plt.title('Titanic Age Distribution')
plt.show()
```

**Kernel Density Distribution**

```python
sns.distplot(ages, hist=True)
plt.xlabel('Ages (years)');
plt.ylabel('Frequency');
plt.title('Titanic Age Distribution')
plt.show()
```

```python
sns.kdeplot(ages)
plt.xlabel('Ages (years)');
plt.ylabel('Frequency');
plt.title('Titanic Age Distribution')
plt.show()
```

```python
sns.kdeplot(ages, bw=10)
plt.xlabel('Ages (years)'); plt.ylabel('Frequency'); plt.title('Titanic Age Distribution')
plt.show()
```

```python
sns.kdeplot(ages, bw=2)
plt.xlabel('Ages (years)'); plt.ylabel('Frequency'); plt.title('Titanic Age Distribution')
plt.show()
```

**Joint Distribution**

```python
sns.jointplot(x='age', y='fare', data=df, kind='scatter')
plt.show()
```
