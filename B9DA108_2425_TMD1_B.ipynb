```python
class Portfolio:
  def __init__(self,name) -> None:
    self.__name=name
    self.__contents =dict()
  def addStock(self,ticker,quantity):
    if ticker in self.__contents:
      self.__contents[ticker]+=quantity
    else:
      self.__contents[ticker]=quantity
  def sellStock(self,ticker,quantity):
    if self.__contents.get(ticker,0)<quantity:
      return False
    if quantity<0:
      raise ValueError('Invalid quantity')
    self.__contents[ticker]-=quantity
    return True
  def value(self,priceList):
    return sum(priceList[ticker]*quantity for ticker,quantity in self.__contents.items())
```

```python
class testPortfolio(unittest.TestCase):
  def testAdd(self):
    p=Portfolio('name')
    p.addStock('AAPL',10)
    self.assertEqual(p.value({'AAPL':180}),1800)
  def testAddMore(self):
    p=Portfolio('name')
    p.addStock('AAPL',10)
    p.addStock('AAPL',10)
    self.assertEqual(p.value({'AAPL':180}),3600)

  def testSell(self):
    p=Portfolio('name')
    p.addStock('AAPL',20)
    p.sellStock('AAPL',10)
    self.assertEqual(p.value({'AAPL':180}),1800)

  def testSellExcess(self):
    p=Portfolio('name')
    p.addStock('AAPL',10)
    p.sellStock('AAPL',100)
    self.assertGreaterEqual(p.value({'AAPL':180}),0)

  def testSellNegative(self):
    p=Portfolio('name')
    p.addStock('AAPL',10)
    with self.assertRaises(ValueError):
      p.sellStock('AAPL',-10)

  def testAddNegative(self):
    p=Portfolio('name')
    p.addStock('AAPL',10)
    with self.assertRaises(ValueError):
      p.addStock('AAPL',-10)

t()
```
