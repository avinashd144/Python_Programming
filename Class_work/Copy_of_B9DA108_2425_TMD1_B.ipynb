```python
#Create a function taking two dictionaries,
#portfolio (id:quantity_int) and prices (id:price_float)
#and computes the value of the portfolio

#add features
 #error correction
 #price file
 #portfolio files (submit a list of filenames for evaluation)


# compute the value of the portfolio
def compute_portfolio_value(portfolio, prices):
  value = 0
  for key in portfolio:
    if key in prices:
      value += portfolio[key] * prices[key]
  return value

portfolio = {
    'A':20,
    'B':30,
    'C':40
    }

prices = {
    'A':85.555,
    'B':75,
    'C':95
    }

total_value = compute_portfolio_value(portfolio, prices)
print(f"The value of portfolio: , {total_value:.3f}")
```

```python
#Re-implement the portfolio problem using a class Portfolio, which should have attributes
#__name : string
#__contents : dict(string:int)

#and provide methods
#addStock
#removeStock
#value

#value should take a priceList : dict(string:float) parameter


class Portfolio:
  def __init__(self, name):
    self.__name = name
    self.__contents = {}

  def addStock(self, stock, units):
    if units <= 0:
      print("units must be a positive integer")
      return
    if stock in self.__contents:
      self.__contents[stock] += units
      print(f"Added {units} units in {stock}")
    else:
      self.__contents[stock] = units
      print(f"Added {units} units in {stock}")

  def removeStock(self, stock, units):
    if units <= 0:
      print("Units must be a positive integer")
      return
    if stock not in self.__contents:
      print("Stock not in portfolio")
    elif self.__contents[stock] < units:
      print(f"Cannot remove {units} units of {stock}, only {self.__contents[stock]} available.")
    else:
      self.__contents[stock] -= units
      print(f"Removed {units} units of {stock} from the portfolio.")
    if self.__contents[stock] == 0:
      print(f"Removed {units} units of {stock} from the portfolio, remaining 0 units")

  def value(self, priceList):
    value = 0
    for stock, quantity in self.__contents.items():
      if stock in priceList:
        value += quantity * priceList[stock]
      else:
        print(f"Warning: Price for {stock} not found in price list.")
    return value

  def __str__(self):
    return f"Portfolio '{self.__name}' contains: {self.__contents}"

my_portfolio = Portfolio("My Portfolio")
my_portfolio.addStock("CGL", 26)
my_portfolio.addStock("BSNL", 36)
my_portfolio.removeStock("CGL", 26)
my_portfolio.addStock("CGL", 26)


# Print portfolio details
print(my_portfolio)

# Define prices
prices = {'CGL': 1636, 'BSNL': 1881}
# Calculate and print total portfolio value
print(f"Total portfolio value: ${my_portfolio.value(prices):.2f}")
```
